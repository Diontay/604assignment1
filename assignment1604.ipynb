# Assignment 1: Understanding your data 

```python
import pandas as pd 
import matplotlib as plt
```

```python
#get carpet data from my github account put in dataframe and print the head 
urlcarpet = 'https://raw.githubusercontent.com/Diontay/604assignment1/main/carpet.csv'
carpet = pd.read_csv(urlcarpet,header=None)
print(carpet.head())
```

```python
# get hardwood data from my github account put in dataframe and print the head 
urlwood = 'https://raw.githubusercontent.com/Diontay/604assignment1/main/hardwood.csv'
wood = pd.read_csv(urlwood,header=None)
print(wood.head())
```

```python
#describe carpet data 
carpet.describe()
```

```python
#describe the wood data 
wood.describe()
```

# statistical information for wood and carpet data 

```python
#number of observations for wood and carpet data 
print(len(wood))
print(len(carpet))
```

```python
#get dimensions of dataframes
print(wood.shape)
print(carpet.shape)
```

```python
#mean of each feature for data frames 
woodMean= wood.mean(axis=0)
carpetMean = carpet.mean(axis=0)

print(woodMean)
print(carpetMean)
```

```python
# median of each feature for both dataframes 
print(wood.median(axis=0))
print(carpet.median(axis=0))
```

```python
#histogram for carpet 
carpet.stack().hist(grid=False,bins = 64)
```

```python
#histogram for hardwood 
wood.stack().hist(grid=False,bins = 64)
```

```python
#scatter plot of features 1 and 54 as a 2d plot for carpet 
carpet.plot.scatter(x = 1, y = 54, c = 'blue')
```

```python
#scatter plot of features 1 and 54 as a 2d plot for hardwood 
wood.plot.scatter(x = 1, y = 54, c = 'blue')
```
